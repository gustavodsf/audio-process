#### Energy

A **energia** ([Wikipedia](https://en.wikipedia.org/wiki/Energy_(signal_processing%29); FMP, p. 66) de um sinal corresponde à magnitude total do sinal. Para sinais de áudio, isso corresponde aproximadamente ao nível de volume do sinal. A energia em um sinal é definida como

$$ \sum_n \left| x(n) \right|^2 $$

A **raiz da energia média quadrática (RMSE)** de um sinal é definido por:

$$ \sqrt{ \frac{1}{N} \sum_n \left| x(n) \right|^2 } $$

* Abertura: Favorita

#### Mel Frequency Cepstral Coefficients (MFCCs)

Calcule o erro entre os tempos de batida estimados e alguma lista de referência de localizações de batidas. Ele apenas transmite um deslocamento constante, ou seja, adicionando um valor constante a todo o espectro. Portanto, muitos praticantes descartam o primeiro MFCC ao realizar a classificação. #### Spectral Centroid

O **centróide espectral** ([Wikipedia](https://en.wikipedia.org/wiki/Spectral_centroid)) indica em qual frequência a energia de um espectro está centrada. Isso é como uma média ponderada:

$$ f_c = \ frac {\ sum_k S (k) f (k)} {\ sum_k S (k)} $$

onde $ S (k) $ é a magnitude espectral na frequência bin $ k $, $ f (k) $ é a frequência na bin $ k $.

* Abertura: Favorita

```python
my_spectral_centroid_plot(audio_fav_intro)
```

* Abertura: Órfões da Terra
